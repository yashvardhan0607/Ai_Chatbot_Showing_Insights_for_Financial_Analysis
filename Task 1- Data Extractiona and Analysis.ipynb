import pandas as pd
df = pd.read_csv(r'D:\JN\financial_data.csv', encoding='ISO-8859-1')
<!-- Importing Pandas and creating a data frame (super table in python) and adding the csv datatset into the data frame -->

```python
df.head()
# displaying the first 5 rows of the table to verify the import
```

df['Total Revenue'] = df['Total Revenue'].replace(r'[\$,]', '', regex=True)
<!-- Removing the $ sign so that it can be converted into the numeric column (Right now it was text) -->

```python
df['Total Revenue'] = pd.to_numeric(df['Total Revenue'], errors='coerce')  
# changing the data type to Numbers from text.
```

```python
df['Revenue Growth %'] = df.groupby(['Company'])['Total Revenue'].pct_change()*100
# Creating a new column in the data frame to represent the Revenue growth in percentage
```

```python
df.head()
```

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
# Creating a new column in the data frame to represent the net income growth in percentage
```

```python
df['Net Income'] = df['Net Income'].replace(r'[\$,]', '', regex=True)
df['Net Income'] = pd.to_numeric(df['Net Income'], errors='coerce')  
```

```python
df.head()
```

```python
df['Net Profit %'] = df['Net Income'] / df['Total Revenue'] * 100
# Evaluating the Net profit %
```

```python
df.head()
```

```python
df['Diluted EPS'] = pd.to_numeric(df['Diluted EPS'], errors='coerce')
# converting the Diluted EPS to numeric
```

```python
df['Diluted EPS Growth (%)'] = df.groupby('Company')['Diluted EPS'].pct_change() * 100
# carring out the % change in Diluted EPS
```

```python
df['ROA (%)'] = (df['Net Income'] / df['Total Assets']) * 100
# Evaluating Return on Assets
```

```python
df.head()
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth %': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()
```

```python
print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```
